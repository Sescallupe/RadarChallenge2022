# **Aumentando datos train**

```python
DA_folder='Aumento_data_train'
DAN_folder='Aumento_data_train_normal'
images_increased = 2

try:
    os.mkdir("/content/drive/MyDrive/Radar_Challenge/Imagenes_Usadas/" + DA_folder)
    os.mkdir("/content/drive/MyDrive/Radar_Challenge/Imagenes_Usadas/" + DAN_folder)
except:
    print("")
```

```python
train_datagen = ImageDataGenerator(
    #rotation_range=3, # realiza rotación ángulo aleatoria
    width_shift_range=0, #costados
    height_shift_range=0, # arriba o abajo
    shear_range=0,  #inclina a las imaganes
    zoom_range=[1,0.15],
    horizontal_flip=True,
    vertical_flip=True)


#data_path = "/content/drive/MyDrive/Radar_Challenge/Imagenes_Utilizadas/Aumento_data_train"
#data_dir_list = os.listdir(data_path)


width_shape, height_shape = 380, 380

i=0
num_images=0
for image_file in data_ruta_train:

    img_path = image_file

    imge = cv2.imread(img_path)
    
    imge=cv2.resize(image.img_to_array(imge), (width_shape, height_shape), interpolation = cv2.INTER_AREA)
    x= imge/255
    x=np.expand_dims(x,axis=0)
    t=1
    for output_batch in train_datagen.flow(x,batch_size=1):
        a=image.img_to_array(output_batch[0])
        imagen1=output_batch[0,:,:]*255
        imagen2=output_batch[0,:,:]
        #imgfinal = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
        name = "_"+image_file.split("/")[7]+"_"+image_file.split("/")[8]
        cv2.imwrite("/content/drive/MyDrive/Radar_Challenge/Imagenes_Usadas/"+DA_folder+"/"+"%s%i%i.jpg"%(name,i,t), imagen1) 
        cv2.imwrite("/content/drive/MyDrive/Radar_Challenge/Imagenes_Usadas/"+DAN_folder+"/"+"%s%i%i.jpg"%(name,i,t), imagen2) 
        t+=1
        
        num_images+=1
        if t>images_increased:
            break
    i+=1
    
print("images generated",num_images)
```
